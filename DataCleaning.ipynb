```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("datasets.csv")
```

```python
df.head()
```

```python
def clean(amt):
    st=''
    amt=amt[2::]
    for i in range(0,len(amt)):
            if amt[i]==',':
                pass
            else:
                st=st+amt[i]
    return st
```

```python
df['hoa'] = df['hoa'].apply(clean)
df['property tax'] = df['property tax'].apply(clean)
df['rent amount'] = df['rent amount'].apply(clean)
df['fire insurance'] = df['fire insurance'].apply(clean)
df['total'] = df['total'].apply(clean)
df.head()
        
```

```python
d= pd.get_dummies(df,columns=['animal','furniture'],drop_first=True)
```
